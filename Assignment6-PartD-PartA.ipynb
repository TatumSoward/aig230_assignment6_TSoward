### D0 - Load and preprocess

```python
# Downloading the corpus
nltk.download('brown')

corpus = nltk.corpus.brown.sents(categories='news')

for sent in corpus[:3]:
    print(sent)
```

```python
# Applying Lowercase
corpus = [[word.lower() for word in sent] for sent in corpus]

for sent in corpus[:3]:
    print(sent)
```

```python
# Removing punctuation but keeping contractions
import re

corpus = [[word for word in sent if re.search(r'\w', word)] for sent in corpus]

for sent in corpus[:3]:
    print(sent)
```

```python
# Adding beginning of sentence and end of sentence (<bos>/<eos>) tokens
corpus = [['<bos>'] + list(sent) + ['<eos>'] for sent in corpus]

for sent in corpus[:3]:
    print(sent)
```

### D1 - Split

```python
# Getting an 80% train, 10% test, 10% validation split
from sklearn.model_selection import train_test_split

train, rest = train_test_split(corpus, train_size = 0.8, shuffle = True) #80% Train
test, val = train_test_split(rest, train_size = 0.5, shuffle = True) #10% Test, 10% Validation

# number of sentences in each split
print("Number of Sentences:")
print(f"Train: {len(train)}")
print(f"Test: {len(test)}")
print(f"Validation: {len(val)}")
print("\n")

# number of tokens in each split
print("Number of Tokens:")
print(f"Train: {sum(len(sent) for sent in train)}")
print(f"Test: {sum(len(sent) for sent in test)}")
print(f"Validation: {sum(len(sent) for sent in val)}")
print("\n")

#size of vocabulary for train
from itertools import chain

vocab = set(chain.from_iterable(train))
print(f"Train Vocab Size: \n{len(vocab)}")
```

### D2 - Vocabulary

```python
# replacing low frequency vocab words with unknown token <unk>
from collections import Counter

train_words = [word for sent in train for word in sent] # getting all word tokens out of their sub-lists
word_count = Counter(train_words) # Counting the words in the training set

# Getting Train Vocab based on frequency
min_freq = 2 # the minimum amount of times that the word can appear to be in the vocabulary
vocab = {word for word, count in word_count.items() if count >= 2} # defining the vocabulary
vocab.add("<unk>")

# Replacing OOV words with <unk>
train = [word if word in vocab else "<unk>" for word in sent for sent in train]
```

```python
# Reporting new values
print(f"Minimum Frequency: {min_freq}")
print(f"Vocabulary Size: {len(vocab)}")
print("\n")

# Top 20 most common words
print("Top 20 Most Common Words")
for word, count in word_count.most_common(20):
    print(f"{word}:{count}")
```

## Part A - Statistical n-gram Language Model

### A1. Train a Trigram Language Model

```python
from nltk.corpus import brown
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import Laplace

# Ensure the data is tokenized words, not raw strings
# Correct format: [['The', 'jury', 'said'], ['The', 'city', 'council']]
train_sentences = brown.sents(categories='news') 

# Re-run the pipeline (n=3 for trigrams)
n = 3
train_data, padded_vocab = padded_everygram_pipeline(n, train_sentences)

# Re-train the model
model = Laplace(n)
model.fit(train_data, padded_vocab)
```

For the trigram model, I used `padded_everygram_pipeline` from `nltk.lm.preprocessing` and I set `n=3`. The smoothing method I used was Laplace (add-1) Smoothing. Smoothing is important because it helps handle OOV n-gram which would be represented with a probability of 0. This is a problem because of how probability is calculated, it can set the probability of the test set to 0, which ruins the model's predictive chances. The ability to provide a less certain answer instead of failing completely is esscential for calculating perplexity, which is how we measure the model's performance.

### A2. Evaluate with Perplexity

```python
from nltk.util import ngrams

#Validation PPL
val_trigrams = [ngram for sent in val for ngram in ngrams(sent, n, 
                 pad_left=True, pad_right=True, 
                 left_pad_symbol="<s>", right_pad_symbol="</s>")]

val_perplexity = model.perplexity(val_trigrams)
print(f"Validation Perplexity: {val_perplexity}")

#Test PPL
test_trigrams = [ngram for sent in test for ngram in ngrams(sent, n, 
                 pad_left=True, pad_right=True, 
                 left_pad_symbol="<s>", right_pad_symbol="</s>")]

test_perplexity = model.perplexity(test_trigrams)
print(f"Test Perplexity: {test_perplexity}")
```

The perplexity score states that the model is choosing between 10,000 words to make a prediction. Which is about what Gen AI predicted. A lower perplexity indicates the more confident the model. What is good about the score is that it is consistent between validation and testing, this shows that the model performance is not dependent on the input data. It also shows that the train-test-validation split was truly random and shows no bias in the split.

Perplexity is affected by several factors. Such as the n-value, the higher the n-value, the more context and words that the model has to choose from, and the lower the perplexity. Vocabulary size and domain, a large vocabulary increases perplexity due to a greater amount of choices for the model and a limited domain also increases perplexity because the model encounters more OOV words and OOV words are equally considered with n-grams when the model is making predictions. Smoothing techniques can affect perplexity, for example Laplace increases perplexity because it adds 1 to the probability calculations to avoid `P=0`, but it also increases the perplexity calculations. Finally, training set size decreases perplexity because it allows for the model to calculate more accurate probabilities.

### A3. Text Generation

```python
def generate_proper_text(model, max_tokens=50):
    # Starting context for a trigram model
    context = ["<s>", "<s>"]
    generated = []

    for _ in range(max_tokens):
        # Sample the next word
        next_word = model.generate(1, text_seed=context)
        
        # Break if we hit the end of a sentence
        if next_word == "</s>":
            break
            
        # Only add actual words to our list
        if next_word not in ["<s>", "</s>"]:
            generated.append(next_word)
            
        # Slide the window: [word1, word2] -> [word2, word3]
        context = (context + [next_word])[-2:]

    return " ".join(generated)

# Print 3 samples of at least 30 tokens
count = 0
while count < 3:
    sample = generate_proper_text(model, max_tokens=60)
    if len(sample.split()) >= 30:
        print(f"Sample {count+1}:\n{sample}\n")
        count += 1
```

The sentences are grammatical, they tend to follow the rules of English instead of being a random collection of words. They are somewhat coherent, they tend to change topics and go on tangents frequently in the sentences. This is due to the trigrams short context span. The limitations include preference to common n-grams/words such as "and", "with", and "which". Which makes the sentences run-on. This is due to Laplace smoothing increasing the probability of these already common n-grams.
